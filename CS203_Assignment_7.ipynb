```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# Define the MLP Model
class MLP_bow(nn.Module):
    def __init__(self):
        super(MLP_bow, self).__init__()
        self.fc1 = nn.Linear(10000, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 128)
        self.fc4 = nn.Linear(128, 64)
        self.fc5 = nn.Linear(64, 2)
        self.dropout = nn.Dropout(0.3)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.dropout(x)
        x = F.relu(self.fc3(x))
        x = self.dropout(x)
        x = F.relu(self.fc4(x))
        x = self.dropout(x)
        x = self.fc5(x)
        return x
```

```python
# Function to count the trainable parameters per layer
def count_parameters_per_layer(model):
    for name, param in model.named_parameters():
        if param.requires_grad:
            print(f"Layer: {name} | Parameters: {param.numel()}")
    total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"\nTotal Trainable Parameters: {total_params}")
```

```python
# Initialize the model
model = MLP_bow()

# Call the function
count_parameters_per_layer(model)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Handle missing values (Replace NaN with an empty string)
x_train = x_train.fillna("")
x_val = x_val.fillna("")
x_test = x_test.fillna("")
x_train_2 = x_train_2.fillna("")
x_val_2 = x_val_2.fillna("")

# Initialize CountVectorizer with max_features=10,000
vectorizer = CountVectorizer(max_features=10000)

# Fit and transform training data
x_train_bow = vectorizer.fit_transform(x_train).toarray()

# Transform validation and test data (DO NOT fit again!)
x_val_bow = vectorizer.transform(x_val).toarray()
x_test_bow = vectorizer.transform(x_test).toarray()
x_train_2_bow = vectorizer.transform(x_train_2).toarray()
x_val_2_bow = vectorizer.transform(x_val_2).toarray()
```

```python
# Print shape of transformed data
print(f"Train BoW Shape: {x_train_bow.shape}")
print(f"Validation BoW Shape: {x_val_bow.shape}")
print(f"Test BoW Shape: {x_test_bow.shape}")
print(f"Train IMDB BoW Shape: {x_train_2_bow.shape}")
print(f"Validation IMBD BoW Shape: {x_val_2_bow.shape}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
```

```python
# Converting NumPy arrays to PyTorch tensors
x_train_tensor = torch.tensor(x_train_bow, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.to_numpy(), dtype=torch.long)
x_val_tensor = torch.tensor(x_val_bow, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val.to_numpy(), dtype=torch.long)
x_test_tensor = torch.tensor(x_test_bow, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.to_numpy(), dtype=torch.long)
x_train_2_tensor = torch.tensor(x_train_2_bow, dtype=torch.float32)
y_train_2_tensor = torch.tensor(y_train_2.to_numpy().astype(int), dtype=torch.long)
x_val_2_tensor = torch.tensor(x_val_2_bow, dtype=torch.float32)
y_val_2_tensor = torch.tensor(y_val_2.to_numpy().astype(int), dtype=torch.long)


# Creating DataLoader for batch processing
batch_size = 32
train_dataset = TensorDataset(x_train_tensor, y_train_tensor)
val_dataset = TensorDataset(x_val_tensor, y_val_tensor)
test_dataset = TensorDataset(x_test_tensor, y_test_tensor)
train_2_dataset = TensorDataset(x_train_2_tensor, y_train_2_tensor)
val_2_dataset = TensorDataset(x_val_2_tensor, y_val_2_tensor)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
train_2_loader = DataLoader(train_2_dataset, batch_size=batch_size, shuffle=True)
val_2_loader = DataLoader(val_2_dataset, batch_size=batch_size, shuffle=False)
```

```python
#Hyperparameters
epochs = 10
learning_rate = 0.0001
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model_bow = MLP_bow()
model_bow.to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model_bow.parameters(), lr=learning_rate)

best_val_acc = 0.0
train_acc_arr = []
val_acc_arr = []
for epoch in range(epochs):
    model_bow.train()
    total_correct, total_samples = 0, 0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model_bow(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        _, predicted = torch.max(outputs, 1)
        total_correct += (predicted == labels).sum().item()
        total_samples += labels.size(0)

    train_acc = total_correct / total_samples
    train_acc_arr.append(train_acc)

    # Validation
    model_bow.eval()
    val_correct, val_samples = 0, 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model_bow(inputs)
            _, predicted = torch.max(outputs, 1)
            val_correct += (predicted == labels).sum().item()
            val_samples += labels.size(0)

    val_acc = val_correct / val_samples
    val_acc_arr.append(val_acc)
    print(f"Epoch [{epoch+1}/{epochs}]: Train Acc = {train_acc:.2f}, Val Acc = {val_acc:.2f}")

    # Save Best Model
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model_bow.state_dict(), "checkpoint_bow.pt")
        print(f"New best model saved with Validation Accuracy: {best_val_acc:.2f}")
```

```python
import torch
from transformers import BertTokenizer, BertModel

class BERTEmbedder:
    def __init__(self, model_name="bert-base-uncased", device=None):
        self.device = device if device else ("cuda" if torch.cuda.is_available() else "cpu")

        # Load tokenizer and model
        self.tokenizer = BertTokenizer.from_pretrained(model_name)
        self.model = BertModel.from_pretrained(model_name).to(self.device)

        # Get the embedding size (768 for BERT)
        self.embedding_size = self.model.config.hidden_size
        self.model_loaded = True

    def get_embeddings(self, texts):
        self.model.eval()  # Set model to evaluation mode
        embeddings_list = []

        with torch.no_grad():  # No gradients needed
            for text in texts:
                # Tokenize the text
                inputs = self.tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
                inputs = {key: val.to(self.device) for key, val in inputs.items()}  # Move to device

                # Get model output
                outputs = self.model(**inputs)

                # Extract the last hidden state and average across tokens for sentence embedding
                embeddings = outputs.last_hidden_state.mean(dim=1)  # Shape: (1, hidden_size)
                embeddings_list.append(embeddings.cpu())  # Move to CPU to avoid memory issues

        return torch.cat(embeddings_list, dim=0)  # Combine all embeddings
```

```python
# Importing library for the metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Finding accuracy, precision, recall and f1 score for the model
accuracy_percentage=(accuracy_score(y_true,y_pred))*100
precision_percentage=(precision_score(y_true, y_pred, average="weighted"))*100
recall_percentage=(recall_score(y_true,y_pred, average="weighted"))*100
f1_percentage=(f1_score(y_true,y_pred, average="weighted"))*100

print("Stats:")
print(f"Accuracy: {accuracy_percentage:.2f}%")
print(f"Precision: {precision_percentage:.2f}%")
print(f"Recall: {recall_percentage:.2f}%")
print(f"F1 Score: {f1_percentage:.2f}%")
```
